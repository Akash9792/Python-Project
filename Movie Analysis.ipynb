```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\\Youtuber.csv')
```

```python
df.dropna(how='all',inplace=True)
```

```python
df.fillna(method='ffill')
```

```python
x=df['Channel Name'].groupby(df.Country)
```

```python
df['Subscribers'] = df['Subscribers'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
```

```python
df
```

```python
df['Average Likes'] = df['Average Likes'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
```

```python
df
```

```python
df['Average Views'] = df['Average Views'].replace({'€': '', ' ': '', 'M': '000000', 'K': '000'}, regex=True).astype(float).astype(int)
```

```python
df
```

```python
df['Average Comments'] = df['Average Comments'].replace({'€': '', ' ': '', 'M': '000000', 'K': '000'}, regex=True).astype(float).astype(int)
```

```python
df
```
